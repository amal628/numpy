Useful for calculating durations.

Alows you to perform operations on data and times like adding  duration to a date.
*   O=objects:Represents python objects
This is a flexible data type that allows for storing any python object within a Numpy array.
*   S=string : Represents byte strings.
Eamples :S1,S2

example: u1,u2.......

# Numpy array copy and vs view

## The difference between copy and view



*   Copy is a new array, and the view is just a view of the original array.
*   copy owns the data any changes made to the copy will not affect original array,
*  the views does not own the data and any changes made to the view will affect the original array.



## copy:

* the copy should not be affected by the changes made to the original array.

1. Make a copy change the original array and display both arrays:

````python
import numpy as np
arr=np.array([1,2,3,4,5])
x=arr.copy()
arr[0]=42
print(arr) # display arr after changing the first element to 42
print(x)   # copy of arr is store
````

### view:
         the view should be affected by the changes mades to the orginal

````python
import  numpy as np
arr=np.array([1,2,3,4,5])
x=arr.view()
arr[0]=42
print(arr)
print(x)
````

### check if array owns its data :
  as mentioned above,copies owns the data and views does not own the data but how can be we check this?
  every numpy array has the attribute base that returns non if the arrays owns the data.
  otherwise the base attribute refers to the original object.

  the copy returns nons.
  the view returns the original array                        

````python
import numpy as np
arr=np.array([1,2,3,4,5])
x=arr.copy()
y=arr.view()
print(x.base)
print(y.base)
````

# creating the filter array

## 1.Create a filter array that will return only values higher than 42:

````python
import numpy as np
arr=np.array([41,42,43,44])
# create an empty list
filter_arr=[]
# go through each element in arr
for element in arr:
  # if the element is higher than 42, set the value to True, otherwise False:
  if element>42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=arr[filter_arr]

print("original array:\n",arr)
print("\nboolean array:\n",filter_arr)
print("\nfiltered new array:\n",newarr)
````

* list 2: [4, 5, 6, 7]

    Without ufunc, we can use Python's built-in zip() method:

````python
x=[1,2,3,4]
y=[4,5,6,7]
z=[]
for i,j in zip(x,y):
  z.append(i+j)
print(z)
````

# NumPy Logs
 #1.logarithms in NumPy:
## NumPy provides three primary functions for calculating logarithms:

* Base 2 logarithm: log2()
* Natural logarithm (Base e): log() or logip() for log(1+x)
* Base 10 logarithm: log10()

* These functions calculate the logarithmic value for each element in the input array.  If a value can't be computed (like taking the log of a negative number or zero), NumPy will return inf or inf for invalid or infinite results, respectively.

# Logs

 NumPy provides functions to perform log at the base 2, e and 10.

 We will also explore how we can take log for any base by creating a custom ufunc

 All of the log functions will place -inf or inf in the elements if the log can not be computed.

 **log at Any Base**

 NumPy does not provide any function to take log at any base, so we can use the **frompyfunc()** function **along with inbuilt function math.log()** with two input parameters and one output parameter.

 If you need to compute logarithms for a base other than 2, 10, or e ,you can use the following: formula:

 image.png

````python
import numpy as np
from math import log

nplog=np.frompyfunc(log,2,1)

print(nplog(100,15))

#here x is100
#here 15 is base
#so we are finding the logarithmic of 100 to base is 15
#ie,log15(100)
````

````python
#can also use this method if its required'
import numpy as np
import math

#custom function to calculate log at my base
def custom_(x,base):
    return math.log(x,base)

#convert the custom function to a ufunc using from numpyfunc()
log_any_base=np.frompyfunc(custom_,2,1)

#example array
arr=np.array([1,10,100,1000])

#calculate log  base 2 for each element
log_base_3=log_any_base(arr,2)

print(log_base_3)

#base to 10

log_base_10=log_any_base(arr,10)
print(log_base_10)
````

# numpy summations

**summations**

what is the difference between summation and addition?
* **addition is done between two arguments whereas summation happens over n elements.**

````python
#add the values in arr1 to the values in arr2:
import numpy as np

arr1=np.array([1,2,3])
arr2=np.array([1,2,3])

newarr=np.add(arr1,arr2)

print(newarr)
````

````python
#sum the values in arr1 to the values in arr2:
import numpy as np

arr1=np.array([1,2,3])
arr2=np.array([1,2,3])

newarr=np.sum([arr1,arr2])

print(newarr)
````

**summation over an axis**
* if you specify axis=1,numpy will sum the returns in each array.

````python
#perform summation in the following array over 1st axis:
import numpy as np

arr1=np.array([1,2,3])
arr2=np.array([1,2,3])

newarr=np.sum([arr1,arr2],axis=1)

print(newarr)
````

* the axis=1 argument specifies that the summation should occur across the second dimension (or alom=ng each row of the 2-d array.)
* for axis+1,numpy sums the elements in each row:
* second row sum+1=2=3=6

**cummulative sum**

* cummulative sum menas partially adding the elements in a array
   * eg the pAartial sum of[1,2,3,4]would be  [1,1+2+3,1+2+3+4]=[1,3,6,10]
   * perform partial sum with the cumsum()function


````python
#perform cummulative summation in the following array:
import numpy as np

arr=np.array([1,2,3])

newarr=np.cumsum(arr)

print(newarr)
````

# numpy differences

**differences**
  
  * a discrete means substracting two successive elements,
       
       * eg for [1,2,3,4] the discrete differences would be [2-1,3-2,4-3]=[1,1,1]
  * to find the discrete differenece,use the diff()function

````python
#compute the discrete difference  of the following of the following array:
import numpy as np

arr=np.array([10,15,25,5])

newarr=np.diff(arr)

print(newarr)
````

* we can perform this operation repeatedly by giving parameter n.
   
   * eg for [1,2,3,4],the discrete difference with n=2 would be [2-1,3-2,4-3]=[1,1,1] then ,**since n=2,we will do it once more with the new result:[1-1,1-1]=[0,0]**

````python
#compute discrete difference of the following array twice:
import numpy as np

arr=np.array([10,15,25,5])

newarr=np.diff(arr,n=2)

print(newarr)
````

# numpy lcm lowest common multiple

**finding lcm(lowest common multiple)**
   * the lowest common multiple is the **smallest number that is a common multiple of two  numbers.**

````python
#find the lcm of the following two numbers:
import numpy as np

num1=4
num2=6

x=np.lcm(num1,num2)

print(x)
````

**returns :12 because that is the lowest common multiple of both numbres (4x3=12and 6x2=12).**

**finding lcm in arrays**

* to find the lowest common multiple of all values in an array,you can use the reduce() method.

* the reduce() method will use the ufunc,in this case the len() function,on each element and reduce the array by one dimension.

````python
#find the lcm of the values of the following array:
import numpy as np

arr=np.array([3,6,9])

x=np.lcm.reduce(arr)

print(x)
````

returns::18 because that is the because that is the lowest common multiple

# numpy set operations

````python
#convert following array with respected elements to a set:
import numpy as np
arr=np.array([1,1,1,2,3,4,5,5,6,7])
x=np.unique(arr)
print(x)
````

**finding union**
* to find the unique values of two arrays,use the **union 1d()** method.

````python
#find union of the following two set arrays:
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
x=np.union1d(arr1,arr2)
print(x)
````

**finding intersection**

* to find only the values that are present in both arrays,use the **intersect1d() method.**

````python
#find intersection of the following twoset arrays:
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
x=np.intersect1d(arr1,arr2,assume_unique=True)
print(x)
````

**note:the intersection1d() method makes an optional arguments assume unique which if set to true can speed up computation.it should always to be set to true when dealing with sets.**

**finding difference**

* to find only the values in the first set that is not present in the second set, use the **setdiff 1d() method.**

````python
#find the difference of the set1 from set2:

import numpy as np

set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])

x=np.setdiff1d(set1,set2,assume_unique=True)

print(x)
````

note:the **setdiff1d() method makes an optional argument assume_unique,**which if set to **true can speed up computation**.it should always be set to true when dealing with sets.

**finding symmetric difference**

* to find only values that are not present in both sets,use the setxor1d() method.

````python
#find the symmetric difference of the set1 and set2:
import numpy as np

set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])

x=np.setxor1d(set1,set2,assume_unique=True)

print(x)
````

note:the **setxor1d() method** takes an optional argument **assume unique**,which if set to **true can speed up computation**.it should always be set to true when dealing with sets.
